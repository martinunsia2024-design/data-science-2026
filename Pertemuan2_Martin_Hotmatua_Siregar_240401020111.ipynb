```python
# Mengimpor library yang dibutuhkan sesuai instruksi modul
import pandas as pd
import numpy as np

# Memuat dataset Titanic langsung dari URL raw GitHub resmi yang valid
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Menampilkan 5 baris pertama untuk memastikan data berhasil dimuat
df.head()
```

### Q1: Berapa jumlah baris dan kolom pada dataset Titanic? Kolom apa saja yang tersedia?

```python
# Mengecek dimensi dataset (baris, kolom)
baris, kolom = df.shape
print(f"Jumlah Baris: {baris}")
print(f"Jumlah Kolom: {kolom}\n")

# Menampilkan daftar nama kolom yang tersedia
print("Daftar kolom yang tersedia:")
print(df.columns.tolist())
```

**Analisis Temuan Q1:**
Berdasarkan hasil eksekusi kode, dataset Titanic ini memiliki data sebanyak 891 baris penumpang dan terdiri dari 12 kolom informasi. Kolom-kolom tersebut merangkum identitas serta kondisi penumpang, seperti PassengerId, Survived (status keselamatan), Pclass (kelas penumpang), Name, Sex, Age, SibSp, Parch, Ticket, Fare, Cabin, dan Embarked.

### Q2: Kolom mana saja yang memiliki missing values? Berapa jumlah dan proporsinya (%) per kolom?

```python
# Menghitung jumlah missing values per kolom
missing_count = df.isnull().sum()

# Menghitung proporsi persentase missing values
missing_percentage = df.isnull().mean() * 100

# Menggabungkan hasil ke dalam sebuah DataFrame agar rapi
missing_table = pd.DataFrame({
    'Jumlah Missing': missing_count,
    'Proporsi (%)': missing_percentage
})

# Menampilkan kolom yang hanya memiliki nilai kosong (lebih besar dari 0)
print(missing_table[missing_table['Jumlah Missing'] > 0].round(2))
```

**Analisis Temuan Q2:**
Ternyata ada 3 kolom di dalam dataset ini yang datanya bolong atau tidak lengkap:
1. Kolom **Age (Usia)** kehilangan 177 data (sekitar 19.87%). Nilai kosong ini nanti perlu diantisipasi jika ingin melakukan analisis mendalam tentang usia.
2. Kolom **Cabin (Nomor Kabin)** memiliki tingkat kerusakan data paling parah, yaitu kehilangan 687 data atau sekitar 77.1%. Karena mayoritas datanya kosong, kolom ini kurang ideal untuk analisis statistik langsung.
3. Kolom **Embarked (Pelabuhan Asal)** hanya kehilangan 2 data saja (0.22%), sehingga kondisinya masih sangat aman.

### Q3: Berapa persen penumpang yang selamat (Survived=1)?

```python
# Menghitung persentase keselamatan menggunakan rata-rata kolom Survived dikali 100
persen_selamat = df['Survived'].mean() * 100
print(f"Persentase penumpang yang selamat: {persen_selamat:.2f}%")
```

**Analisis Temuan Q3:**
Dari total keseluruhan penumpang yang tercatat di dataset ini, hanya ada sekitar 38.38% penumpang yang berhasil selamat dari tragedi tenggelamnya kapal Titanic. Sisanya, yaitu sekitar 61.62% penumpang, dinyatakan tidak selamat.

### Q4: Filter penumpang wanita dari kelas 1 (Pclass=1). Berapa jumlahnya dan berapa rata-rata usianya?

```python
# Memfilter data menggunakan boolean indexing untuk wanita di kelas 1
wanita_kelas1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]

# Menghitung jumlah dan rata-rata usianya
jumlah_wanita_k1 = len(wanita_kelas1)
rata_usia_wanita_k1 = wanita_kelas1['Age'].mean()

print(f"Jumlah penumpang wanita kelas 1: {jumlah_wanita_k1} orang")
print(f"Rata-rata usia mereka: {rata_usia_wanita_k1:.1f} tahun")
```

**Analisis Temuan Q4:**
Ketika kita menyaring data khusus untuk kelompok penumpang wanita yang berada di kelas 1, kita menemukan ada total 94 orang. Kelompok wanita kelas elite ini memiliki rata-rata usia sekitar 34.6 tahun.

### Q5: Gunakan groupby untuk menghitung tingkat keselamatan (mean Survived) per kelas penumpang. Kelas mana yang paling tinggi?

```python
# Mengelompokkan data berdasarkan Pclass dan menghitung rata-rata keselamatan
tingkat_selamat_per_kelas = df.groupby('Pclass')['Survived'].mean() * 100
print(tingkat_selamat_per_kelas.round(2))
```

**Analisis Temuan Q5:**
Melalui pengelompokkan data berdasarkan kelas tiket (Pclass), terlihat kontras perbedaan peluang keselamatan yang sangat jelas:
* Kelas 1 memiliki tingkat keselamatan tertinggi mencapai **62.96%**.
* Kelas 2 berada di angka **47.28%**.
* Kelas 3 memiliki tingkat keselamatan paling rendah, yaitu hanya **24.24%**.

Temuan ini membuktikan adanya prioritas penyelamatan bagi penumpang kelas atas saat evakuasi darurat berlangsung di kapal.

### Q6: Analisis Tambahan: Apakah tarif tiket (Fare) yang dibayar penumpang berpengaruh terhadap peluang keselamatan mereka?

```python
# Membandingkan rata-rata tarif tiket (Fare) antara kelompok yang selamat dan tidak selamat
rata_tarif_keselamatan = df.groupby('Survived')['Fare'].mean()

print("Rata-rata tarif tiket berdasarkan status keselamatan:")
print(f"Tidak Selamat (0) : ${rata_tarif_keselamatan[0]:.2f}")
print(f"Selamat (1)       : ${rata_tarif_keselamatan[1]:.2f}")
```

**Analisis Temuan Q6:**
Sebagai analisis tambahan, saya membedah korelasi antara tarif tiket (Fare) dengan keselamatan. Hasilnya menunjukkan tren yang sangat menarik. Penumpang yang tidak selamat rata-rata hanya membayar tiket seharga $22.12. Sementara itu, penumpang yang selamat rata-rata membeli tiket seharga $48.40 (dua kali lipat lebih mahal). Hal ini semakin memperkuat kesimpulan sebelumnya bahwa faktor ekonomi atau kelas akomodasi sangat berpengaruh besar pada peluang bertahan hidup di tragedi Titanic.
