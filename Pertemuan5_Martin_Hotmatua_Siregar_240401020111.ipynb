## Bagian 1: Pembuatan Grafik Dasar Menggunakan Matplotlib
Pada tahap awal ini, kita akan menggunakan pustaka `matplotlib.pyplot` untuk memvisualisasikan data tren perkembangan jumlah penumpang transportasi per bulan sepanjang tahun (Line Chart) dan perbandingan performa antar-kategori produk (Bar Chart) menggunakan pendekatan komponen Figure dan Axes.

```python
# Mengimpor pustaka inti visualisasi data sesuai modul praktikum
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Menyiapkan data tren tiruan untuk visualisasi Line Chart
bulan = ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Agu', 'Sep', 'Okt', 'Nov', 'Des']
penumpang_kota_a = [1200, 1350, 1410, 1580, 1620, 1750, 1900, 1850, 1690, 1550, 1420, 1650]

# Membuat visualisasi menggunakan pendekatan object-oriented (Figure & Axes)
fig, ax = plt.subplots(figsize=(8, 4.5))

# Memplot grafik garis dengan kustomisasi penanda (marker) dan warna
ax.plot(bulan, penumpang_kota_a, marker='o', linewidth=2, color='#028090', label='Kota A')

# Menambahkan kelengkapan informasi grafik agar mudah dibaca
ax.set_title('Tren Jumlah Penumpang Bulanan Transportasi (2024)', fontsize=13, pad=12)
ax.set_xlabel('Bulan', fontsize=11)
ax.set_ylabel('Jumlah Penumpang (Orang)', fontsize=11)
ax.grid(True, linestyle='--', alpha=0.6)
ax.legend()

# Menampilkan hasil grafik dasar Matplotlib
plt.tight_layout()
plt.show()
```

**Analisis Grafik Bagian 1 (Pendekatan What? So what? Now what?):**
* **What? (Apa yang diukur?):** Grafik garis di atas mengukur volume jumlah penumpang bulanan di Kota A sepanjang tahun 2024, yang bergerak di rentang nilai 1.200 hingga 1.900 orang per bulan.
* **So what? (Apa maknanya?):** Terlihat ada tren kenaikan yang konsisten dari awal tahun dan mencapai puncak tertinggi (*peak season*) pada bulan Juli (1.900 orang). Pola ini kemungkinan besar dipengaruhi oleh masa libur sekolah dan libur pertengahan tahun. Setelah Juli, angka penumpang kembali melandai secara bertahap.
* **Now what? (Apa tindak lanjutnya?):** Manajemen transportasi Kota A disarankan untuk meningkatkan kuota armada dan mengoptimalkan jadwal penugasan pengemudi khusus pada bulan Juni hingga Agustus guna mengantisipasi lonjakan kepadatan penumpang.

## Bagian 2: Pembuatan Visualisasi Statistik Menggunakan Seaborn
Sekarang kita beralih ke pustaka Seaborn untuk membuat visualisasi tingkat lanjut yang kaya akan informasi statistik. Kita akan memuat dataset multi-variabel bawaan, yaitu `penguins`, untuk membedah sebaran karakteristik fisik hewan serta hubungan korelasi antarfiturnya menggunakan visualisasi *Pair Plot*.

```python
# Memuat dataset multi-variabel 'penguins' yang disediakan oleh Seaborn
df_penguins = sns.load_dataset('penguins')

# Menampilkan informasi ringkas dataset untuk memahami fitur yang tersedia
print("=== EKSPLORASI DATASET PENGUINS ===")
print(f"Dimensi Data: {df_penguins.shape}")
print(df_penguins[['species', 'island', 'bill_length_mm', 'flipper_length_mm', 'body_mass_g']].head())

# Mengatur tema visualisasi Seaborn agar terlihat elegan
sns.set_theme(style="ticks")

# Membuat Pair Plot untuk melihat distribusi univariat dan scatter plot bivariat sekaligus
# Memisahkan kelompok data berdasarkan variabel kategori 'species' (Spesies Penguin)
print("\nMemproses visualisasi Pair Plot (Mohon tunggu beberapa saat)...")
grid_plot = sns.pairplot(df_penguins, hue='species', palette='Set2', corner=True)

# Menampilkan visualisasi matriks statistik Seaborn
plt.show()
```

**Analisis Grafik Bagian 2 (Pendekatan What? So what? Now what?):**
* **What? (Apa yang diukur?):** Matriks visualisasi *pair plot* di atas memetakan sebaran distribusi frekuensi variabel numerik (panjang paruh, lebar paruh, panjang sirip, berat tubuh) sekaligus diagram pencar (*scatter plot*) hubungan antarpasangan fitur tersebut berdasarkan tiga jenis spesies penguin (Adelie, Chinstrap, Gentoo).
* **So what? (Apa maknanya?):** Grafik menampilkan pengelompokan alami (*clustering*) yang sangat tegas. Spesies penguin **Gentoo** (warna hijau/tergantung palet) terbukti memiliki karakteristik fisik yang paling bongsor karena mendominasi nilai tertinggi pada visualisasi sebaran berat tubuh (`body_mass_g`) dan panjang sirip (`flipper_length_mm`). Sementara itu, spesies **Adelie** dan **Chinstrap** memiliki ukuran tubuh mirip, tetapi dapat dibedakan dengan jelas lewat fitur panjang paruh (`bill_length_mm`) mereka.
* **Now what? (Apa tindak lanjutnya?):** Karena variabel panjang paruh dan panjang sirip memiliki pola pemisah kelompok yang sangat baik dan tidak saling bertumpuk, kedua fitur fisik ini wajib dipilih sebagai variabel prediktor utama jika kita ingin membangun model kecerdasan buatan (*Machine Learning*) untuk mengklasifikasikan jenis spesies penguin secara otomatis di masa mendatang.
